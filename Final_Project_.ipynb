# Rainsford "An Historical Account of the Black Empire of Hayti"

First we will open Rainsford's text and prepare it to be read by our Natural Language Processor, SpaCy.

```python
with open("Rainsford_Full_Text.txt", "r", encoding="utf-8") as infile:
  R_full_text = infile.read()

print(f"Download complete. Total characters: {len(R_full_text)}")
print("-" * 50)
print(R_full_text[0:1000])
```

```python
print("Processing text pipeline... (this may take a moment)")
R_doc_full_text = nlp(R_full_text)

print("Processing complete!")
print(f"Processed {len(R_full_text)} characters from the full text.")
```

### Quantifying Mentions of Female-related Terms in Rainsford

To quantify female-related terms, I will create a list of gendered terms related to women and namnameing the variable "female_terms". I then will create a table listing all moments these tokens are mentioned. This way, I can procure a count of the number of times each gender term is mentioned. I found that the easiest way to achieve this is through creating a loop. This way, the code will continuously parse through the text and look for words that will match my list, which will be be counted using "matching_female_tokens"

To create the dataframe, I will use a series using "matching_female tokens" as my data source. Modeling the Visualizing History notebook, I will use ".reset_index" to change my index into a column, making sure to name the column 'Counts'. However, in order for the table to generate correctly, I need to use ".value_counts()" similar to the bar graph generated in NER with SpaCy. This allows the code to understand that the index is the unique tokens and the counts are their unique values.

Then, I will give my table a name and print the top 20 instances of female related words.This will be repeated for each gendered repository and text.

```python
female_terms = [
    'woman', 'women', 'female', 'females', 'girl', 'girls',
    'lady', 'ladies', 'mother', 'mothers', 'daughter', 'daughters',
    'wife', 'wives', 'sister', 'sisters'
]

R_matching_female_tokens = []
for token in R_doc_full_text:
    if token.text.lower() in female_terms:
        R_matching_female_tokens.append(token.text.lower())
    elif token.lemma_.lower() in female_terms:
        R_matching_female_tokens.append(token.lemma_.lower())

R_df_female_mentions = pd.Series(R_matching_female_tokens).value_counts().reset_index(name='Counts')

print("Top mentions of female-related terms:")
display(R_df_female_mentions.head(20))
```

```python
print(f"Sum of all Rainsford female-related terms: {R_df_female_mentions['Counts'].sum()}")
```

The table shows that mother and wife were the two most mentioned female-related terms in the text. These two terms dominate over half of the overall counts listed.
Now let's generate the counts of male-related terms in Rainsford's text using the same process, but different variables and terms.

### Quantifying Mentions of Male-related Terms in Rainsford

```python
male_terms = [
    'man', 'men', 'male', 'males', 'boy', 'boys',
    'father', 'fathers', 'son', 'sons',
    'husband', 'husbands', 'brother', 'brothers'
]

R_matching_male_tokens = []
for token in R_doc_full_text:
    if token.text.lower() in male_terms:
        R_matching_male_tokens.append(token.text.lower())
    elif token.lemma_.lower() in male_terms:
        R_matching_male_tokens.append(token.lemma_.lower())

R_df_male_mentions = pd.Series(R_matching_male_tokens).value_counts().reset_index(name='Counts')

print("Top mentions of male-related terms:")
display(R_df_male_mentions.head(20))
```

Just from the percentages, we can see a disparity in the way women and men are refered to in Rainsford's text. Women are refered to using relational terms 64% of the time compared to men being refered to by relational terms 32% of the time.

But, is this signifigant? I will use a chi-square test to determine the signifigance of use of relational terms for each gender. I will be modeling the chi-square test shown in the in-class jupyter notebook "python for statistics". The chi-square table will allow me to test the null hypothesis that there is no relation between gender and relational terms. 

To run this command, I will need to install a new package called scipy. Scipy allows for a special command called 'chi2_contingency' to be used to help streamline creating a chi-square table. To download this, I had to place '%' in front of the command to activate it in a notebook cell! I learned this from following a guide VS Code provides during a SyntaxError.

```python
%pip install scipy

from scipy.stats import chi2_contingency
```

Next, I simply inserted the corosponding values for each variable. To get the nonrelational-terms, I subtracted the rational terms sum by the total number of gendered terms for both female and male words. Then, I ensured I would have each of my 4 variables represented by labling the index 'relation' and 'nonrelation' and the columns 'female' and 'male'.

```python
female_relational = 84
female_nonrelational = 47
male_relational = 79
male_nonrelational = 170

table = pd.DataFrame(
    [[female_relational, male_relational],
     [female_nonrelational, male_nonrelational]],
    index=['relation', 'nonrelational'],
    columns=['female', 'male']
)
display(table)

chi2, p, dof, expected = chi2_contingency(table)
print("chi2:", chi2)
print("p-value:", p)
print("expected:", expected)
```

Awesome! A p-value < 0.05 means that the difference between the values is unlikely due to chance. In Rainsford's text, we can see that the p = ~0.00000026. This *much* less than 0.05. and is a sign to reject the null hypothesis. The null hypothesis was that the relationship between gender and relational-terms is independent

In addition, the expected values show up what the values should be if gender and relational-terms were independant. If these two veriables were independant, then relational female terms should be 56.19 instead of 84. Female nonrelations terms should be 74.81, not 47. Looking at language used for males, male relational terms were expected to be 106.81, not 79. Male nonrelational terms were expected to be 142.19 instead of 170.

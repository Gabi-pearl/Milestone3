```python
import spacy
import pandas as pd

try:
      nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading language model...")
    from spacy.cli import download
    download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

print("✅ Libraries imported and Model loaded!")
```

```python
with open("Edwards_Full_Text.txt", "r", encoding="utf-8") as infile:
  E_full_text = infile.read()

print(f"Download complete. Total characters: {len(E_full_text)}")
print("-" * 50)
print(E_full_text[0:500])
```

```python
print("Processing text pipeline... (this may take a moment)")
E_doc_full_text = nlp(E_full_text)

print("Processing complete!")
print(f"Processed {len(E_full_text)} characters from the full text.")
```

```python
female_terms = [
    'woman', 'women', 'female', 'females', 'girl', 'girls',
    'lady', 'ladies', 'mother', 'mothers', 'daughter', 'daughters',
    'wife', 'wives', 'sister', 'sisters'
]

E_matching_female_tokens = []
for token in E_doc_full_text:
    if token.text.lower() in female_terms:
        E_matching_female_tokens.append(token.text.lower())
    elif token.lemma_.lower() in female_terms:
        E_matching_female_tokens.append(token.lemma_.lower())
```

```python
E_df_female_mentions = pd.Series(E_matching_female_tokens).value_counts().reset_index(name='Counts')

print("Top mentions of female-related terms:")
display(E_df_female_mentions.head(20))
```

```python
male_terms = [
    'man', 'men', 'male', 'males', 'boy', 'boys',
    'father', 'fathers', 'son', 'sons',
    'husband', 'husbands', 'brother', 'brothers'
]

E_matching_male_tokens = []
for token in E_doc_full_text:
    if token.text.lower() in male_terms:
        E_matching_male_tokens.append(token.text.lower())
    elif token.lemma_.lower() in male_terms:
        E_matching_male_tokens.append(token.lemma_.lower())

E_df_male_mentions = pd.Series(E_matching_male_tokens).value_counts().reset_index(name='Counts')

print("Top mentions of male-related terms:")
display(E_df_male_mentions.head(20))
```
